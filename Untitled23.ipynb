**Part 2: Explore & Clean Data**

**Part 3: Compute Statistics**

```python
# Part 3: Compute Statistics

# List of numerical columns to analyze
columns = ['price', 'area', 'bedrooms']

for col in columns:
    print(f"Statistics for {col}:\n")
    print(f"Mean: {df[col].mean()}")
    print(f"Median: {df[col].median()}")
    print(f"Mode: {df[col].mode()[0]}")
    print(f"Variance: {df[col].var()}")
    print(f"Standard Deviation: {df[col].std()}")
    print(f"Skewness: {df[col].skew()}")
    print(f"Kurtosis: {df[col].kurt()}")
    print(f"Minimum: {df[col].min()}")
    print(f"Maximum: {df[col].max()}\n")
```

**Part 4: Optional Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Part 4: Optional Visualization

# Histograms for Price and Area
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['price'], bins=30, kde=True)
plt.title('Histogram of Price')

plt.subplot(1, 2, 2)
sns.histplot(df['area'], bins=30, kde=True)
plt.title('Histogram of Area')

plt.tight_layout()
plt.show()

# Box plots for Price and Area
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(x=df['price'])
plt.title('Box Plot of Price')

plt.subplot(1, 2, 2)
sns.boxplot(x=df['area'])
plt.title('Box Plot of Area')

plt.tight_layout()
plt.show()

# Scatter plot of Price vs Area
plt.figure(figsize=(8, 6))
sns.scatterplot(x='area', y='price', data=df)
plt.title('Price vs Area')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```
